## Task 1b: Load Data

Load the Iris data (same as in the Pandas tasks).


```python
df = pd.read_csv('iris.csv')
df.head(2)
```

## Task 2a: Create a figure

Create the a scatter plot using the iris data, but compare `petal_width` to `petal_length`.  Be sure to add labels to the axes.

```python
# create plot
fig = plt.figure()
ax = fig.add_subplot(1 ,1, 1)
ax.scatter(df['petal_length'].values, df['petal_width'].values)

# add axes labels
ax.axes.set_xlabel('Petal Length')
ax.axes.set_ylabel('Petal Width')

# add figure title
ax.set_title('Petal length vs width')

# show plot w labels
plt.show()
```

## Task 3a: Figure Style

Create a scatter plots using the iris data frame. Do not convert the data frame columns to numpy arrays. Rather use the dataframe as is. Be sure to:
- plot `petal_length` vs `petal_width`.
- Use column names to specify the x and y coordinates (don't pass in Numpy arrays)
- Color each dot by the species (you'll need to be creative to do this).

```python
df['species'].unique()
```

```python
# create plot
fig = plt.figure()
ax = fig.add_subplot()

# set color abbreviations
color = df['species'].replace(
    ['setosa', 'versicolor', 'virginica'], 
    ['b', 'g', 'm'])

ax.scatter(df['petal_length'], df['petal_width'], c=color)

# add axes labels
ax.axes.set_xlabel('Petal Length')
ax.axes.set_ylabel('Petal Width')

# add figure title
ax.set_title('Petal length vs width')

# show plot w labels
plt.show()
```

## Task 4a: Categorical plots

Create a bar graph that indicates the mean `sepal_length` of each subspecies.

```python
# create plot
bar_graph = plt.figure(figsize=(10, 5))
bg_ax = bar_graph.add_subplot(1,3,1)

#group df by species, get mean of each measurement
group = df.groupby('species').mean()

# set groups and values
bg_ax.bar(group.index, group['sepal_length'])

# Add a title to the figure.
bg_ax.set_title('Mean sepal length by sub spp')

# Cleanly show the plot.
plt.show()
```

## Task 6a:  Multiple Plots

Create 4 scatterplots for the iris data all in one figure. The figure should include scatterplots of these comparisions:
- `sepal_length` x `petal_length`
- `sepal_length` x `petal_width`
- `sepal_width` x `petal_length`
- `sepal_width` x `petal_width`

Be sure that:
- the points for the species 
- the axes for each plot are labeled
- the figure has a label.

```python
# figure object
fig2 = plt.figure(figsize=(10,10))

# set color abbreviations
color = df['species'].replace(
    ['setosa', 'versicolor', 'virginica'], 
    ['b', 'g', 'm'])

# make scatter plots
axes = fig2.subplots(2, 2)
axes[0,0].scatter(df['sepal_length'], df['petal_length'], c=color)
axes[0,1].scatter(df['sepal_length'], df['petal_width'], c=color)
axes[1,0].scatter(df['sepal_width'], df['petal_length'], c=color)
axes[1,1].scatter(df['sepal_width'], df['petal_width'], c=color)

# make labels and titles
fig2.suptitle('Measurement Comparisons, color by sub spp')
axes[0,0].set_title('sepal length vs petal length')
axes[0,1].set_title('sepal_length vs petal_width')
axes[1,0].set_title('sepal_width vs petal_length')
axes[1,1].set_title('sepal_width vs petal_width')


plt.show()
```
